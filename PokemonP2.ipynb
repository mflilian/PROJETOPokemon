## **Projeto Pokemón** - Análise de Correlação


Objetivo: Analisar atributos dos Pokémons através do uso de graficos, dados estatísticos e análise de correlações a fim de criar uma relação com a base de dados obtida na primeira parte do projeto.

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


pd.read_csv('http://professor.ufabc.edu.br/~ronaldo.prati/bases/Pokemon_sim.csv')
df = pd.read_csv('http://professor.ufabc.edu.br/~ronaldo.prati/bases/Pokemon_sim.csv')

df = df.drop(columns='Legendary')
df
```

```
#Ranqueamento por vitórias

df = df.sort_values(by='vitorias', ascending=False) 
df
```

explicar q vou separar entre type 1 type 2 vitorias derrotas e empates e o pq de eu fazer isso





```
df = df[["Type 1" , "Type 2" , "vitorias" , "derrotas" , "empates"]]
df

```

analisando vitorias e tipo de pokemon 1

```
fig, ax = plt.subplots(figsize=(12,7))
ax.set_title("Vitorias x Tipo(1) do Pokemon", fontdict={'fontsize':15})
sns.barplot(x=df['Type 1'], y=df['vitorias'], ax=ax, data=df, palette='RdPu').plot(kind='bar',rot=60)
ax.annotate("Maior valor", 
            xy=(1, 50000),
            color= 'red',
            size= '16',
            xytext=(1, 50000))
fig.tight_layout()
```

```
fig, ax = plt.subplots(figsize=(12,7))
ax.set_title("Empates x Tipo(1) do Pokemon", fontdict={'fontsize':15})
sns.barplot(x=df['Type 1'], y=df['empates'], ax=ax, data=df, palette='RdPu').plot(kind='bar',rot=60)
ax.annotate("Maior valor", 
            xy=(11, 600),
            color= 'red',
            size= '16',
            xytext=(11, 600))
fig.tight_layout()
```

```
fig, ax = plt.subplots(figsize=(12,7))
ax.set_title("Vitorias x Tipo(2) do Pokemon", fontdict={'fontsize':15})
sns.barplot(x=df['Type 2'], y=df['vitorias'], ax=ax, data=df, palette='RdPu').plot(kind='bar',rot=60)
ax.annotate("Maior valor", 
            xy=(3, 50000),
            color= 'red',
            size= '16',
            xytext=(3, 50000))
fig.tight_layout()
```

```
fig, ax = plt.subplots(figsize=(12,7))
ax.set_title("Derrotas x Tipo(2) do Pokemon", fontdict={'fontsize':15})
sns.barplot(x=df['Type 2'], y=df['derrotas'], ax=ax, data=df, palette='RdPu').plot(kind='bar',rot=60)
ax.annotate("Menor valor", 
            xy=(3, 28000),
            color= 'red',
            size= '16',
            xytext=(3, 28000))
fig.tight_layout()
```
