## **Projeto Pokemón**

Objetivo: Analisar alguns atributos dos Pokémons através do uso de graficos, dados estatísticos e análise de correlações e relacionar com a primeira parte do projeto



```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Essa é a tabela onde encontram-se todos os dados, uma tabela mais completa que a da primeira parte, com simulções de batalhas e outras informções

```

df= pd.read_csv("http://professor.ufabc.edu.br/~ronaldo.prati/bases/Pokemon_sim.csv")
df = df.drop(columns='partidas')
df[(df['Legendary'] == False)]
```

Assim como na 1 parte do prejeto, começamos calculando os valores e organizando em colunas e gráficos. Mas como essa tabela dispõe de mais valores e simulções, faremos comparações mais complexas.
Começando com o agrupamento dos valos Máximos de Rank conforme o tipo de pokemons.

```
p=df.tail(150)
p[(p['Legendary'] == False)]

z=p.groupby("Type 1")["rank"].max()*100
z
```

Nesta célula analisaremos os tipos de pokemons que apresentam os indivíduos mais fortes, ou seja, aquele tipo de pokemon que têm o maior rank em relação aos demais. Como podemos obeservar, O Tipo Fada e Dark são aqueles que detem um pokemon cujo rank é superior aos demais. Vale ressaltar que, como o Rank representa as Vitórias daquele pokemon de 0 a 1, multiplicamos por 100 para comparar os valores

```
p=df.tail(150)
p[(p['Legendary'] == False)]

z=p.groupby("Type 1")["rank"].max()*100
plt.figure(figsize=(12,5))
plt.title("Rank máximo x Tipo do pokemon")
plt.ylabel("Rank máximo")
plt.xlabel("Type 1")
plt.grid()
y=plt.plot(z)
y
```

Agora, vamos relizar o cálculo da mediana e usar esses valores para construir um gráfico posteriormente. (Assim como no anterior, á fins comparativos, multiplicamos os valores por 100)

```
x=df.tail(150)
x[(x['Legendary'] == False)]
mediana= x.groupby("Type 1")["rank"].median()*100
mediana
```

Aqui realizamos e criamos um gráfico conforme a mediana dos valores do Rank(Vitórias) conforme cada tipo dos Pokemons. Aqui, temos outro indicativo de que os tipos ''Rock'' e ''Dragon'' têm Pokemons mais fortes que os demais, dessa vez evidênciado pela mediana. 

```
y3=x.groupby('Type 1')['rank'].median()*100

plt.figure(figsize=(13,5))
plt.ylabel("Mediana (Rank)")
plt.xlabel("Type 1")
plt.grid()
plt.plot(y3)
```

Depois disso, compararemos 2 informações: A média de vitórias e o Rank máximo. Nesse sentido, é possível obsevar que os tipos de pokemons que possuem alguns Pokemons extremamente fortes, nem sempre têm a maior média, como no caso do tipo 'Fary', que atinge um maximo alto no rank, mas média baixa. 

```
x=df.tail(150)
x[(x['Legendary'] == False)]


y1=x.groupby("Type 1")["rank"].max()*100
y2=x.groupby("Type 1")["% vitorias"].mean()
y3=x.groupby('Type 1')['rank'].median()*100

plt.figure(figsize=(13,5))
plt.title("Rank máximo x Tipo do pokemon")
plt.ylabel("Rank máximo, Vitórias")
plt.xlabel("Type 1")
plt.grid()

plt.plot(y1, label="Rank Máximo")
plt.plot(y2, label="Média de Vitórias")
plt.plot(y3, label="Mediana")

plt.legend()
plt.show()
```
